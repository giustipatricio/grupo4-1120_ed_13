```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import preprocessing
import re
```

```python
df = pd.read_csv(r'./properati_train.csv')
```

```python
df.state_name.unique()
```

```python
data = df[['price_usd_per_m2','state_name','property_type','surface_total_in_m2','new_rooms_clean','place_name_clean','surface_total_in_m2_clean','state_name']]
```

```python
data['log_price'] = np.log(data['price_usd_per_m2'])
data['log_superficie'] = np.log(data['surface_total_in_m2_clean'])
```

```python
print(min(data['price_usd_per_m2']),max(data['price_usd_per_m2']))
```

```python
data.price_usd_per_m2.quantile([0.25,0.5,0.75])
```

```python
lista = []
for i in data['price_usd_per_m2']:
    if i <=1800:
        lista.append(str('bajo'))
    elif i> 1800 and i<= 2486:
        
        lista.append(str('medio'))
    else:
        i > 2486.41
        lista.append(str('alto'))
    
    
```

```python
data['barrios_category'] = lista
```

```python
data = pd.get_dummies(data = data, columns = ['barrios_category','property_type'], drop_first=True)
```

```python
data.log_price.dropna(inplace=True)
```

```python
data.dropna(subset=['log_price'],inplace=True)
```

```python
data1 = data.query('state_name == "Capital Federal"')
```

```python
data1 = pd.get_dummies(data = data1, columns = ['place_name_clean','new_rooms_clean'], drop_first=True)
```

```python
X_caba = data1[['log_superficie','barrios_category_bajo','barrios_category_medio', 'property_type_apartment',
       'property_type_house', 'property_type_store', 'place_name_clean_Agronomía', 'place_name_clean_Almagro',
       'place_name_clean_Balvanera', 'place_name_clean_Barracas',
       'place_name_clean_Barrio Norte', 'place_name_clean_Belgrano',
       'place_name_clean_Boca', 'place_name_clean_Boedo',
       'place_name_clean_Caballito', 'place_name_clean_Capital Federal',
       'place_name_clean_Catalinas', 'place_name_clean_Centro / Microcentro',
       'place_name_clean_Chacarita', 'place_name_clean_Coghlan',
       'place_name_clean_Colegiales', 'place_name_clean_Congreso',
       'place_name_clean_Constitución', 'place_name_clean_Flores',
       'place_name_clean_Floresta', 'place_name_clean_Las Cañitas',
       'place_name_clean_Liniers', 'place_name_clean_Mataderos',
       'place_name_clean_Monserrat', 'place_name_clean_Monte Castro',
       'place_name_clean_Nuñez', 'place_name_clean_Once',
       'place_name_clean_Palermo', 'place_name_clean_Palermo Chico',
       'place_name_clean_Palermo Hollywood', 'place_name_clean_Palermo Soho',
       'place_name_clean_Palermo Viejo', 'place_name_clean_Parque Avellaneda',
       'place_name_clean_Parque Centenario',
       'place_name_clean_Parque Chacabuco', 'place_name_clean_Parque Chas',
       'place_name_clean_Parque Patricios', 'place_name_clean_Paternal',
       'place_name_clean_Pompeya', 'place_name_clean_Puerto Madero',
       'place_name_clean_Recoleta', 'place_name_clean_Retiro',
       'place_name_clean_Saavedra', 'place_name_clean_San Cristobal',
       'place_name_clean_San Nicolás', 'place_name_clean_San Telmo',
       'place_name_clean_Tribunales', 'place_name_clean_Velez Sarsfield',
       'place_name_clean_Versalles', 'place_name_clean_Villa Crespo',
       'place_name_clean_Villa Devoto', 'place_name_clean_Villa General Mitre',
       'place_name_clean_Villa Lugano', 'place_name_clean_Villa Luro',
       'place_name_clean_Villa Ortuzar', 'place_name_clean_Villa Pueyrredón',
       'place_name_clean_Villa Real', 'place_name_clean_Villa Riachuelo',
       'place_name_clean_Villa Santa Rita', 'place_name_clean_Villa Soldati',
       'place_name_clean_Villa Urquiza', 'place_name_clean_Villa del Parque','new_rooms_clean_1.0',
        'new_rooms_clean_2.0', 'new_rooms_clean_3.0',
       'new_rooms_clean_4.0', 'new_rooms_clean_5.0', 'new_rooms_clean_6.0']]
y_caba = data1['log_price']

#spliteamos en set de testeo y entrenamiento
X_train, X_test, y_train, y_test = train_test_split(X_caba, y_caba, test_size=0.30, random_state=10)
```

```python
lm = linear_model.LinearRegression()

model_1 = lm.fit(X_train, y_train)

y_pred = lm.predict(X_test)

print ('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print ('MSE:', metrics.mean_squared_error(y_test, y_pred))
print ('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print ('R2:', metrics.r2_score(y_test, y_pred))
print('Score model_1: test', model_1.score(X_test, y_test))
print('Score model_1: train', model_1.score(X_train, y_train))
```

```python
lm_lasso = linear_model.LassoCV(alphas=[0.00001, 0.00005, 0.0001, 0.0005, 0.001, 0.005, 0.01,\
                                        0.05, 0.1, 1, 5, 10, 15, 25, 50,100,200],\
                                        normalize=True, cv=10)

model_3 = lm_lasso.fit(X_train, y_train)
y_pred = lm_lasso.predict(X_test)

print ('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print ('MSE:', metrics.mean_squared_error(y_test, y_pred))
print ('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print ('R2:', metrics.r2_score(y_test, y_pred))
print('Score model_3:', model_3.score(X_test,y_test))
```

```python
lm_ridge = linear_model.RidgeCV(alphas=[0.00001, 0.00005, 0.0001, 0.0005, 0.001, 0.005, 0.01,\
                                        0.05, 0.1, 1, 5, 10, 25, 50 ],\
                                        normalize=True, cv=5) 
# Definimos el rango de de búsqueda del hiperparametro explicitamente

model_2 = lm_ridge.fit(X_train, y_train)
y_pred = lm_ridge.predict(X_test)
print ('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print ('MSE:', metrics.mean_squared_error(y_test, y_pred))
print ('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print ('R2:', metrics.r2_score(y_test, y_pred))
print('Score model_2 test:', model_2.score(X_test, y_test))
print('Score model_2 train:', model_2.score(X_train, y_train))
```
